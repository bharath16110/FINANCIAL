```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#total bill vs tip datasets
tips=sns.load_dataset('tips')
```

```python
#seaborn using distplot()
sns.distplot(tips['total_bill'])
plt.show()
```

```python
#seaborn using swarmplot()
sns.swarmplot(x="day",y="total_bill",data=tips)
plt.show()
```

```python
#seaborn using stipplot
sns.stripplot(x="day",y="total_bill",data=tips)
plt.show()
```

```python
#seaborn using displot()
sns.displot(tips['total_bill'])
plt.show()
```

```python
#seaborn using pairplot()
sns.set_style("ticks")
sns.pairplot(tips)
plt.show()
```

```python
#seaborn using distplot()
sns.distplot(tips['total_bill'],hist=0)
plt.show()
```
